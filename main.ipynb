```python
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings("ignore")
import pickle as pkl
```

```python
df = pd.read_csv('./Music Info.csv')
df 
```

```python
df.columns
```

```python
df = df[['track_id', 'name', 'artist', 'spotify_preview_url', 'spotify_id', 'tags']]
df 
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['tags'] = df['tags'] + df['artist']
```

```python
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
vector = cv.fit_transform(df['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
similarty = cosine_similarity(vector)
```

```python
def recommend(song):
    song_lower = song.lower()
    
    if not any(df['name'].str.lower() == song_lower):
        print(f"Error: The song '{song}' does not exist in the dataset.")
        return
    
    index = df[df['name'].str.lower() == song_lower].index[0]
    
    distance = sorted(list(enumerate(similarty[index])), reverse=True, key=lambda vector: vector[1])
    
    print(f"Recommendations for '{song}':")
    for i in distance[0:6]:  
        print(df.iloc[i[0]]['name'])
        print('link', df.iloc[i[0]]['spotify_preview_url'])
        print('artist', df.iloc[i[0]]['artist'])
```

```python
recommend('Thunder')
```

```python
pkl.dump(df, open('songs.pkl', 'wb'))
```

```python
pkl.dump(similarty, open('similarity.pkl', 'wb'))
```
